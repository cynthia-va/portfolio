```python
import numpy as np
import pandas as pd

pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb

from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
```

```python
#Read dataset
data_filepath = 'bank-marketing-cleandata.csv'
data = pd.read_csv(data_filepath, delimiter = ',', encoding = 'utf-8')

data.head()
```

```python
data.shape
```

```python
#Variables conversion from non-numeric to numeric
d = defaultdict(LabelEncoder)

def fit_and_transform (column) :
    if data[column.name].dtype == 'object' :
        return d[column.name].fit_transform(column)
    else :
        return column
fit = data.apply(fit_and_transform)
```

```python
fit.head()
```

```python
#Check correlations
correlations = fit.corr()

plt.figure(figsize = (16,14))
sb.heatmap(data = correlations, annot = True)
```
